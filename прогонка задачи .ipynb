```python
import numpy as np
import matplotlib.pyplot as plt

# Constants for the domain and mesh
a = 0
b = 1
N = 100
h = (b-a)/N

# Boundary conditions
gamma0 = 1
delta0 = 0
gamma1 = 1
delta1 = 0
A = 1
B = np.exp(-1) + 1  # e^-1 + 1

# Functions according to the differential equation
def p(x):
    return x - 1

def q(x):
    return 3.125

def f(x):
    return 4 * x

# Sweep method for tridiagonal systems
def sweep(ai, bi, ci, di):
    alpha = np.zeros(N)
    beta = np.zeros(N)
    y = np.zeros(N+1)
    
    # Forward sweep
    alpha[0] = -delta0 / gamma0
    beta[0] = A / gamma0
    for i in range(1, N):
        alpha[i] = ci[i-1] / (bi[i-1] - ai[i-1]*alpha[i-1])
        beta[i] = (ai[i-1]*beta[i-1] - di[i-1]) / (bi[i-1] - ai[i-1]*alpha[i-1])
    
    # Backward substitution
    y[N] = (B - beta[N-1]*delta1) / (gamma1 + alpha[N-1]*delta1)
    for i in range(N-1, -1, -1):
        y[i] = alpha[i] * y[i+1] + beta[i]
    
    return y

# Setup coefficients for the tridiagonal matrix
ai = np.zeros(N-1)
bi = np.zeros(N-1)
ci = np.zeros(N-1)
di = np.zeros(N)

# Initialize coefficients
for i in range(1, N):
    x_i = a + i * h
    ai[i-1] = 1.0 - p(x_i) * h / 2.0
    bi[i-1] = -2.0 - q(x_i) * h * h
    ci[i-1] = 1.0 + p(x_i) * h / 2.0
    di[i-1] = f(x_i) * h * h

di[0] = f(a) * h * h + (1 - p(a) * h / 2) * A
di[N-1] = f(b) * h * h + (1 + p(b) * h / 2) * B

# Compute the solution
yy = sweep(ai, bi, ci, di)

# Print each x and y value
for i in range(N+1):
    print('x = {:.2f}, y[{}] = {:.4f}'.format(a + i * h, i, yy[i]))

# Plot the numerical solution
x_values = np.linspace(a, b, N + 1)
plt.plot(x_values, yy, label='Numerical Solution')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Numerical Solution of the Differential Equation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

def integrate(F, x, y, xStop, h):
    def run_kut4(F, x, y, h):
        K0 = h * F(x, y)
        K1 = h * F(x + h / 2.0, y + K0 / 2.0)
        K2 = h * F(x + h / 2.0, y + K1 / 2.0)
        K3 = h * F(x + h, y + K2)
        return (K0 + 2.0 * K1 + 2.0 * K2 + K3) / 6.0
    
    X = []
    Y = []
    X.append(x)
    Y.append(y)
    while x < xStop:
        h = min(h, xStop - x)
        y = y + run_kut4(F, x, y, h)
        x = x + h
        X.append(x)
        Y.append(y)
    return np.array(X), np.array(Y)

def F(x, y):
    F = np.zeros(2)
    F[0] = y[1]
    F[1] = -3.0 * y[0] * y[1]
    return F

def initCond(u):
    return np.array([u, 0])

def printSoln(X, Y, freq):
    for i in range(0, len(X)):
        if i % freq == 0:
            print(f"x = {X[i]:.2f}, y = {Y[i][0]:.4f}")

def r(u):
    X, Y = integrate(F, xStart, initCond(u), xStop, h)
    y = Y[-1]
    r = y[0] - 1.0
    return r

def bisection(f, a, b, tol=1e-6):
    fa, fb = f(a), f(b)
    if fa * fb >= 0:
        raise Exception("Function values at the endpoints must have different signs")
    while (b - a) / 2 > tol:
        c = (a + b) / 2
        fc = f(c)
        if fc == 0:
            return c
        elif fc * fa < 0:
            b = c
            fb = fc
        else:
            a = c
            fa = fc
    return (a + b) / 2

# Main simulation parameters
xStart = 0.0   # Start of integration
xStop = 2.0    # End of integration
u1 = 1.0       # 1st trial value of unknown init. cond.
u2 = 2.0       # 2nd trial value of unknown init. cond.
h = 0.1        # Step size
freq = 2       # Printout frequency

# Compute the correct initial condition using the bisection method
u = bisection(r, u1, u2)

# Perform integration with the correct initial condition
X, Y = integrate(F, xStart, initCond(u), xStop, h)

# Print results
printSoln(X, Y, freq)
```

```python
import math
from numpy import sign
import sys
import numpy as np
import matplotlib.pyplot as plt

def err(string):
    print(string)
    input('Press return to exit')
    sys.exit()

def ridder(f, a, b, tol=1.0e-9):
    points_x = [a, b]
    points_y = [f(a), f(b)]

    fa = f(a)
    fb = f(b)
    if fa == 0.0:
        plot_results(points_x, points_y, a)
        return a
    if fb == 0.0:
        plot_results(points_x, points_y, b)
        return b
    if sign(fa) == sign(fb):
        err("Function has the same signs at the endpoints a and b")
    
    xOld = a  # начальное значение для проверки сходимости
    for i in range(30):
        c = 0.5 * (a + b)
        fc = f(c)
        points_x.append(c)
        points_y.append(fc)
        
        s = math.sqrt(fc**2 - fa*fb)
        if s == 0.0:
            err("Zero division error in calculating s")

        dx = (c - a) * fc / s
        if (fa - fb) < 0.0:
            dx = -dx
        x = c + dx
        fx = f(x)
        points_x.append(x)
        points_y.append(fx)

        print(f"Iteration {i+1}:")
        print(f"a = {a}, b = {b}, c = {c}, x = {x}")
        print(f"fa = {fa}, fb = {fb}, fc = {fc}, fx = {fx}")
        print(f"dx = {dx}, s = {s}\n")

        if i > 0 and abs(x - xOld) < tol * max(abs(x), 1.0):
            plot_results(points_x, points_y, x)
            return x
        xOld = x

        if sign(fc) == sign(fx):
            if sign(fa) != sign(fx):
                b = x
                fb = fx
            else:
                a = x
                fa = fx
        else:
            a = c
            b = x
            fa = fc
            fb = fx
    
    err("Too many iterations")

def plot_results(points_x, points_y, root):
    plt.figure(figsize=(10, 5))
    plt.scatter(points_x, points_y, color='red', label='Function values')
    plt.plot(points_x, points_y, label='Function curve')
    plt.scatter([root], [0], color='blue', s=100, label='Root found')
    plt.title('Ridder\'s Method Root Finding')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.axhline(0, color='gray', lw=0.5)
    plt.legend()
    plt.grid(True)
    plt.show()

# Пример использования:
def test_function(x):
    return x**2 - 3

root = ridder(test_function, 1, 2)
print("Root is:", root)
```
